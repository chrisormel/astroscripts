```python
import HO23
import numpy as np
import pylab as pl
```

```python
#figure 2 and 13 of OH23
qp = 3e-6
eta = 3e-3
h = 0.05 #used for Bondi radius
rb = qp/h**2 #Bondi radius-over-orbital radius
rh = (qp/3)**(1/3) #Hill radius

Rlist = [1.0*rh,0.1*rh,0.01*rh,1e-3*rh,1e-4*rh,rb]

npt = 100
tauarr = 10**np.linspace(0,4,npt)

fg, ax = pl.subplots(1,1)

for k,rcap in enumerate(Rlist):
    if rcap==rb:
        col='orange'; lw=2.0; lstyle='--'
    else:
        col='k'; lw=1.0; lstyle='-'
        
    epsarr, Stplat, Stcrit, Stres = HO23.epsilon_HO (tauarr, qp, eta, rcap)
    ax.loglog(tauarr,epsarr, ls=lstyle, c=col, lw=lw)

    ax.set_xlabel('Stokes number')

ax.axhline(1.0, ls='--', c='0.7', lw=2.0)
ax.set_ylim(1e-3,1.1)
ax.set_ylabel('accretion probability, $\epsilon$')
```
