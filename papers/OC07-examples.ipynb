```python
import numpy as np
import OC07
import pylab as pl
import matplotlib.cm as cm

npt = 200
Starr = 10**np.linspace(-6,4,npt)

St1, St2 = np.meshgrid(Starr,Starr)

#with and without finite inertial range (Re=1e8)
#cf. Ormel & Cuzzi (2007), figs. B and C
Delv1 = OC07.Dv_turb(St1,St2)
Delv2 = OC07.Dv_turb(St1,St2,Re=1e8)

fg, (ax1,ax2) = pl.subplots(1,2, figsize=(10,4))

levL = [1e-4,3e-4,1e-3,3e-3,0.01,0.03,0.1,0.3,0.8,1,1.15]


CS1 = ax1.contour(Starr, Starr, Delv1,levels=levL,cmap=cm.rainbow)
CS2 = ax2.contour(Starr, Starr, Delv2,levels=levL,cmap=cm.rainbow)

cbar = fg.colorbar(CS1)
cbar.ax.set_yscale('linear')

for ax in [ax1,ax2]:
    ax.set_xscale('log')
    ax.set_yscale('log')
    ax.set_aspect('equal')
    ax.set_xlabel('Stokes number')
ax1.set_ylabel('Stokes number')

cbar.ax.set_ylabel('relative velocity [$V_g$]')
```
